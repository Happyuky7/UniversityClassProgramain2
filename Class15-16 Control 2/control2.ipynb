**Campana de Gauss | Desviación estándar:**

Podemos apreciar la desviación estándar representados en un gráfico de la Campana de Gauss `(1)`.

Formula distribución normal con la campana de gauss `(2)`:



Referencia:
- `(1)` Definicion.de. _Definición de Campana de Gauss._ Recuperado el 12 de septiembre, 2024, de https://definicion.de/campana-de-gauss/
- `(2)` Matematicasonline.es. _DISTRIBUCIÓN NORMAL o campana de Gauss-Laplace._ Recuperado el 12 de septiembre, 2024, de https://www.matematicasonline.es/BachilleratoCCSS/segundo/archivos/distribucion_normal/DISTRIBUCION%20NORMAL.htm

```python
x = np.linspace(min(pf['passengers']), max(pf['passengers']), 1000)

# Calcular la distribución normal (Campana de Gauss)
y = (1 / (desviacion * np.sqrt(2 * np.pi))) * np.exp(-0.5 * ((x - media) / desviacion) ** 2)

plt.figure(figsize=(10, 6))
plt.plot(x, y, label='Distribución Normal', color='blue')
plt.axvline(x=media, color='r', linestyle='--', label='Media')
plt.axvline(x=mediana, color='purple', linestyle='--', label='Mediana')
plt.axvline(x=media + desviacion, color='g', linestyle='--', label='Media + 1 Desviación Estándar')
plt.axvline(x=media - desviacion, color='g', linestyle='--', label='Media - 1 Desviación Estándar')

plt.fill_between(x, y, where=((x >= media - desviacion) & (x <= media + desviacion)), color='green', alpha=0.3, label='Área bajo la curva (1 Desviación Estándar)')

plt.title('Campana de Gauss de la cantidad de pasajeros')
plt.xlabel('Cantidad de pasajeros')
plt.ylabel('Densidad de probabilidad')
plt.legend(title='Simbología:')
plt.grid(True)
plt.show()
```

Se puede deducir que desde la media esta tiene una dispersión es de aproximandamente +- 100 pasajeros.

**Gráfico Media**

```python
plt.figure(figsize=(10, 6))
plt.plot(data['year'], data['passengers'], label='Pasajeros', marker='o')
plt.axhline(y=media, color='r', linestyle='--', label='Media de pasajeros')
plt.title('Cantidad de pasajeros respecto a la media')
plt.xlabel('Año')
plt.ylabel('Cantidad de pasajeros')
plt.legend(title='Simbología:')
plt.grid(True)
plt.show()
```

```python
x = np.linspace(min(pf['passengers']), max(pf['passengers']), 1000)

# Calcular la distribución normal (Campana de Gauss)
y = (1 / (desviacion_15 * np.sqrt(2 * np.pi))) * np.exp(-0.5 * ((x - media_15) / desviacion_15) ** 2)

plt.figure(figsize=(10, 6))
plt.plot(x, y, label='Distribución Normal', color='blue')
plt.axvline(x=media_15, color='r', linestyle='--', label='Media')
plt.axvline(x=mediana_15, color='purple', linestyle='--', label='Mediana')
plt.axvline(x=media_15 + desviacion_15, color='g', linestyle='--', label='Media + 1 Desviación Estándar')
plt.axvline(x=media_15 - desviacion_15, color='g', linestyle='--', label='Media - 1 Desviación Estándar')

plt.fill_between(x, y, where=((x >= media_15 - desviacion_15) & (x <= media_15 + desviacion_15)), color='green', alpha=0.3, label='Área bajo la curva (1 Desviación Estándar)')

plt.title('Campana de Gauss de la cantidad de pasajeros')
plt.xlabel('Cantidad de pasajeros')
plt.ylabel('Densidad de probabilidad')
plt.legend(title='Simbología:', loc='upper right', bbox_to_anchor=(1.3, 1))
plt.grid(True)
plt.show()
```

Se puede deducir que desde la media esta tiene una dispersión es de aproximandamente +- 200 pasajeros.

Si comparamos estos datos con la anterior gráfica, observamos claramente que en este caso su dispersión es mayor. Ademas que su media y mediana se transladaron hacia la derecha.
